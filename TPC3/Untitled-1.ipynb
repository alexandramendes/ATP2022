```python
print("Bem vindo!")

Menu = ('Menu\nEscolha a opção:\n(1)Criar Lista;\n(2)Ler Lista;\n(3)Soma;\n(4)Média;\n(5)Maior;\n(6)Menor;\n(7)Ordenada por ordem crescente;\n(8)Ordenada por ordem decrescente;\n(9)Procura um elemento;\n(0)Sair')

Lista = []

print(Menu)

print("ATENÇÃO!\nSe escolher qualquer opção maior que 2, irá ter que executar a opção 1 ou 2 primeiro!")

programa = True
while programa == True:
    n = int(input("escolha uma opção n"))
    if n == 0:
        print("Obrigada por jogares!\nAté à próxima :)")
        programa = False
    if n == 1:
        import random
        for i in range (1,101):
            Lista.append(random.randint(1,100))
            Lista.sort()
        print(Lista)
    if n == 2:
        Lista.clear()
        print("Escolha quantos elementos quer que a sua lista tenha!")
        tamanho = int(input("Escolha quantos elementos quer que a sua lista tenha!"))
        while tamanho > 0:
            x = int(input("Escolha os elementos da sua lista"))
            Lista.append(x)
            tamanho = tamanho - 1
        print(Lista)
    if n ==3:
        soma = 0
        for i in Lista:
            soma = soma + i
        print(soma)
    if n == 4:
        soma = 0
        for i in Lista:
            soma = soma + i
            media = soma / len(Lista)
        print(media)
    if n == 5:
        max = Lista[0]
        for i in Lista[1:]:
            if i > max:
                max = i
        print(max)
    if n == 6:
        men = Lista[0]
        for i in Lista[1:]:
            if i < men:
                men = i
        print(men)
    if n == 7:
        listacopiada = Lista.copy ()
        listacopiada.sort ()
        if listacopiada == Lista:
            print("Sim")
        else:
            print("Não")
    if n == 8:
        listacopiada = Lista.copy ()
        listacopiada.sort (reverse = True)
        if listacopiada == Lista:
            print("Sim")
        else:
            print("Não")
    if n == 9:
        a = int(input("Escolha o elemento que procura"))
        if a in Lista:
            print("A posição do número é {}".format(Lista.index(a)))
        else:
            print("O elemento não está na lista")
```
