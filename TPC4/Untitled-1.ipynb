```python
cinema = []

def inserirsala(cinema):
    filme = input("Introduza o nome do filme:")
    for p in cinema:
        nlugares, vendidos, nome = p
        if nome == filme:
            print("Filme já existe!")
            menu()
    nlugares = int(input("Introduza o número de lugares da sala:"))
    vendidos = []
    cinema.append((nlugares, vendidos, filme))
    print("Sala criada com sucesso!")
    print("Salas:", cinema)
    menu()



def listar(cinema):
    print("Sala Filme")
    print("------------------------")
    for i in cinema:
        nlugares, vendidos, filme = i
        print("Sala#{0}".format(cinema.index(i)), " ", filme)
    menu()

def disponivel( cinema, filme, lugar ):
    disp = True
    for i in cinema:
         nlugares, vendidos, nome = i
         if filme == nome:
            if lugar in vendidos:
                disp = False
    return disp

def vendebilhete (cinema, filme, lugar):
    if disponivel( cinema, filme, lugar):
        for i in cinema:
            nlugares, vendidos, nome = i
            if filme == nome:
                vendidos.append(lugar)
        print("Lugar vendido com sucesso!")
        print("Salas:",cinema)
    else:
        print("Lugar indisponivel!")
    menu()

def listardisponibilidades(cinema):
    print("Sala   Filme   Lugares Disponíveis ")
    print("-------------------------")
    for i in cinema:
        nlugares, vendidos, filme = i
        print("Sala#{0}".format(cinema.index(i)), " ", filme, " ", nlugares - len(vendidos))
    menu()



def menu():
    print("--------------------------\n(1)Criar sala\n(2)Listar salas\n(3)Lugar disponivel\n(4)Vender bilhetes\n(5)Consultar salas\n(0)Sair/Reset\n---------------------")
    menuoption = int(input("Escolha uma opçao:"))
    if menuoption == 1:
        inserirsala(cinema)
    elif menuoption == 2:
        listar(cinema)
    elif menuoption == 3:
        filme = input("Introduza o nome do filme:")
        lugar = int(input("Introduza o lugar:"))
        print("Lugar disponivel?", disponivel(cinema, filme, lugar))
    elif menuoption == 4:
        filme = input("Introduza o nome do filme:")
        lugar = int(input("Introduza o lugar:"))
        vendebilhete(cinema, filme, lugar)
    elif menuoption == 5:
        listardisponibilidades(cinema)
    elif menuoption == 0:
        print("Reset efetuado com sucesso!")
        print("Salas:", cinema)
menu()
```
